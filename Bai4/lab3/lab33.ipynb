```python
def get_geometric_mean_for_three_numbers(a, b, c): 
    return (a*b*c)**(1/2)
```

```python
from functools import reduce  


def get_geometric_mean(*nums: float) -> float: 
    """ 
    Get the geometric mean of a sequence of numbers  
    """ 

    if not len(nums):  
        raise ValueError("Cannot calculate the geometric mean of an empty sequence") 

    product = reduce(lambda a, b: a * b, nums) 
    if product < 0 and len(nums) % 2 == 0: 
        raise ValueError("Cannot calculate the geometric mean") 

    return pow(product, 1 / len(nums)) 
```

```python
print(get_geometric_mean_for_three_numbers(2,4,8))
print(get_geometric_mean(2,4,8))
print(get_geometric_mean(1,3,9))
```

This code defines two functions to calculate the geometric mean.

The function get_geometric_mean_for_three_numbers calculates the geometric mean of three numbers by multiplying them and taking the square root of the product.

The function get_geometric_mean is more flexible because it accepts any number of inputs using *nums. It multiplies all numbers using reduce and then calculates the n-th root of the product.

The function also includes error handling for empty sequences and invalid calculations with negative numbers.
